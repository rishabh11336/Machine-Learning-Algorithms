## Sample problem of predicting per capita income in Canada

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

#### **Problem Statement:** With given data build a machine learning model that can predict per capita income in Canada by year  
- Lets remember equation from high school $$ y = mx+b $$

```python
from IPython.display import Image

# Load the image
image = Image(filename="/content/Income.png")

# Display the image in the text cell
display(image)
```

```python
df = pd.read_csv("/content/canada_per_capita_income.csv")
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
```

```python
y = df["per capita income (US$)"]
X = df.drop(["per capita income (US$)"], axis=1)
```

## Let's Train Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X,y)
```

#### Y = m * X + b (m is coefficient and b is intercept)

```python
clf.coef_
```

```python
clf.intercept_
```

#### Let's find the per capita in year 2000 and 2014

```python
data = {"year":[2000,2014]}
test = pd.DataFrame(data)
```

```python
pred = clf.predict(test)
pred
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
plt.plot(df.year, clf.predict(df[['year']]), color='blue')
```
